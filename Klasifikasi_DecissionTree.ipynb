```python
#import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

# Decision tree

```python
#load dataset
data = pd.read_csv('K01_heart_disease.csv')
data
```

# Preprocessing Data 

```python
data_list_string = list()
for col in data:
    data_list_string.append(data[col].unique().tolist())
data_list_string
```

```python
X = data.drop('HeartDisease',axis=1)
Y = data['HeartDisease']
```

# Encoding Data String To int

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encode = LabelEncoder()
for col in X:
    X[col] = encode.fit_transform(X[col])
Y = encode.fit_transform(Y)
```

```python
data_list_num = list()
for col in X:
    data_list_num.append(X[col].unique().tolist())
data_list_num
```

```python
data_key = {}
for i in range(len(data_list_num)):
    for j in range(len(data_list_num[i])):
        data_key[data_list_string[i][j]] = data_list_num[i][j]
data_key
```

# Modeling 

```python
from sklearn.tree import DecisionTreeClassifier
model_tree = DecisionTreeClassifier()
model_tree.fit(X,Y)
```

```python
data_test = [50,'M','NAP',140,200,0,'Normal',160,'N',1.0,'Up']
for i in range(len(data_test)):
    data_test[i] = data_key[data_test[i]]
data_test
```

```python
prediksi = model_tree.predict([data_test])
if prediksi == 2:
    print("Hasil prediksi play = yes")
else:
    print("Hasil prediksi play = no")
```
